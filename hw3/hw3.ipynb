```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
sns.set_style("whitegrid")
sns.set_palette(sns.color_palette("Greens", 1))
```

```python
# Risk free return
rf = 0.02
# Supposing equally weighted market 
betas = np.asarray([0.8,1,1.2]) 
returns = np.asarray([.1,.1,.14])
rm = (betas*returns).mean()
x = np.arange(0,2, .1)
y = rf + x*(rm-rf)
```

```python
ax = sns.lineplot(x=x,y=y, label="SML for Beta Weighted Portfolio")
ax.set_title('Security Market Line'), ax.set_xlabel('Beta'), ax.set_ylabel('Required Return')
l = sns.scatterplot(x=[0.8, 1, 1.2],y=[.1, .1, .14], color='black')
pts = [(0.8, .1), (1, .1), (1.2, .14)]
for pt, lab in zip(pts, ['A', 'B', 'C']):
    l.text(pt[0], pt[1], lab,horizontalalignment='left', 
     size='medium', color='black', weight='semibold')
plt.show()
```
